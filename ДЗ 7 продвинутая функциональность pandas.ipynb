```python
import pandas as pd
```

```python
contact_preference = pd.read_csv('contact_preference.csv', sep=';')
customer_preferenced_contact_type = pd.read_csv('customer_preferenced_contact_type.csv', sep=';')
customer = pd.read_csv('customer.csv', sep=';')
customers_segmetns = pd.read_csv('customers_segmetns.csv', sep=';')
location = pd.read_csv('location.csv', sep=';')
product = pd.read_csv('product.csv', sep=';')
product_feedback  = pd.read_csv('product_feedback.csv', sep=';')
purchase = pd.read_csv('purchase.csv', sep=';')
segment = pd.read_csv('segment.csv', sep=';')
service_feedback = pd.read_csv('service_feedback.csv', sep=';')
```

```python
df = customer.merge(location, how ='left', on='location_id')

df = df.merge(customers_segmetns, how ='left', left_on='id', right_on = 'customer_id').drop('customer_id', axis=1)

df = df.merge(segment, how ='left', left_on='segment_id', right_on = 'id').drop('id_y', axis=1)
df.rename(columns={'id_x': 'id'}, inplace=True)

df = df.merge(customer_preferenced_contact_type, how ='left', left_on='id', right_on = 'customer_id').drop('customer_id', axis=1) 

df = df.merge(contact_preference, how ='left', left_on='contact_type_id', right_on = 'contact_id').drop('contact_id', axis=1)

df = df.merge(service_feedback, how ='left', left_on='id', right_on = 'customer_id').drop('customer_id', axis=1)
df.rename(columns={'id_y': 'service_feedback_id'}, inplace=True)
df.rename(columns={'id_x': 'customer_id'}, inplace=True)

user_data = df.merge(purchase, left_on='customer_id', right_on='user_id', how='left', suffixes=('', '_purchase'))   

user_data = user_data.merge(product, left_on='product_id', right_on='id', how='left', suffixes=('', '_product'))

user_data = user_data.merge(product_feedback, left_on='id', right_on='purchase_id',how='left', suffixes=('', '_productfeedback'))
user_data.drop('purchase_id', axis=1)
user_data.head(15)
```

# Задание 1

```python
user_data['amount'] = user_data['quantity'] * user_data['price']
user_data['category_name'] = user_data['category_name'].fillna('')
user_data.head(15)
```

```python
def most_popular_category(x):
    return x.value_counts().index[0]

user_summary = user_data.groupby('customer_id').agg(
    cnt_purchase = ('id','count'),
    total_amount = ('amount','sum'),
    avg_raiting = ('rating_productfeedback','mean'),
    max_raiting = ('rating_productfeedback','max'),
    most_popular_product_category=('category_name',most_popular_category),
    cnt_review = ('id_productfeedback','count')
)

user_summary
```

# Задание 2

```python
def calculate_spending_share(user_id):
    user_purchases = user_data[user_data['customer_id'] == user_id]
    
    total_spent = user_data['amount'].sum()

    category_spending = user_data[user_data['category_name'].isin(['Movies', 'Games', 'Books'])].groupby('category_name')['amount'].sum().reset_index()
    
    category_spending['share'] = category_spending['amount'] / total_spent
    
    return category_spending

spending_share = calculate_spending_share(1)
spending_share
```
